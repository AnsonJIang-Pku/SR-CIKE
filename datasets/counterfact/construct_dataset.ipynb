#### 分割原始数据集，10k,6k,4k

```python
# 将counterfact分割为train、dev和test 三个数据集，比例为 10000:6000:4000
import pandas as pd
import numpy as np

random_seed = 42

data_dir = '/home/zjiang/repos/TODO/CrossLingual_IKE/IKE_base/counterfact/original/counterfact_cleaned.json'
train_dir = '/home/zjiang/repos/TODO/CrossLingual_IKE/IKE_base/counterfact/train/counterfact_train.json'
dev_dir = '/home/zjiang/repos/TODO/CrossLingual_IKE/IKE_base/counterfact/dev/counterfact_dev.json'
test_dir = '/home/zjiang/repos/TODO/CrossLingual_IKE/IKE_base/counterfact/test/counterfact_test.json'
# 1. 加载JSON文件
df = pd.read_json(data_dir)  # 如果JSON是每行一条记录，可以添加 lines=True

# 2. 检查数据量是否足够
total_samples = len(df)
required_samples = 10000 + 6000 + 4000
if total_samples < required_samples:
    raise ValueError(f"数据不足，需要至少{required_samples}条，但只有{total_samples}条")

# 3. 随机打乱数据
df = df.sample(frac=1, random_state=random_seed).reset_index(drop=True)

# 4. 分割数据集
train = df.iloc[:10000]
dev = df.iloc[10000:16000]  # 10000+6000=16000
test = df.iloc[16000:20000]  # 16000+4000=20000

# 5. 验证分割是否正确
print(f"Train集大小: {len(train)}")
print(f"Dev集大小: {len(dev)}")
print(f"Test集大小: {len(test)}")

# 6. 保存为新的JSON文件
train.to_json(train_dir, orient='records', indent=4)
dev.to_json(dev_dir, orient='records', indent=4)
test.to_json(test_dir, orient='records', indent=4)
```
